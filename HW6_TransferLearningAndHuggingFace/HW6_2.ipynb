```python
from transformers import pipeline, AutoImageProcessor#, AutoModelForTextToImage
from PIL import Image
import requests
from diffusers import StableDiffusionPipeline


# Initialize translation pipeline (Chinese to English)
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-zh-en")

# Initialize text-to-image pipeline using StableDiffusionPipeline
text_to_image = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
# If using GPU, move the pipeline to the device
# text_to_image = text_to_image.to("cuda") # Uncomment if using GPU
```

```python
# Task 1: Translate Chinese to English
chinese_text = input("\u8f38\u5165\u4e2d\u6587\u53e5\u5b50 (Enter a Chinese sentence): ")
translated = translator(chinese_text, max_length=512)
english_text = translated[0]['translation_text']
print(f"\u7ffb\u8b6f\u7d50\u679c (Translated sentence): {english_text}")

# Task 2: Generate image from translated text
print("\u5728\u6a21\u578b\u4e2d\u751f\u6210\u5716\u50cf... (Generating image...)")
# Generate and save the image directly
image = text_to_image(english_text, num_inference_steps=50, guidance_scale=7.5).images[0]
image.save(f"generated_image.png")
image.show()
```
